```python
crop_dict = {
    'rice' : 1 ,
    'maize' : 2,
    'chickpea' : 3,
    'kidneybeans' : 4,
    'pigeonpeas': 5,
    'mothbeans': 6,
    'mungbean' : 7,
    'blackgram' : 8,
    'lentil' : 9,
    'pomegranate': 10,
    'banana' : 11,
    'mango' : 12,
    'grapes' : 13,
    'watermelon' : 14,
    'muskmelon' : 15,
    'apple' : 16,
    'orange' : 17,
    'papaya' : 18,
    'coconut' : 19,
    'cotton' :  20,
    'jute' : 21,
    'coffee' : 22
}
```

```python
crop['crop_no'] = crop['label'].map(crop_dict)
crop.head()
```

```python
crop.tail()
```

```python
crop.drop('label', axis=1, inplace=True)
crop.tail()
```

# Train Test Split

```python
x = crop.drop('crop_no',axis=1)
y = crop['crop_no']

print(f"The shape of x is : {x.shape}")
print(f"The shape of y is : {y.shape}")
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
print(f"Shape of x_train: {x_train.shape}")
print(f"Shape of y_train: {y_train.shape}")
print(f"Shape of x_test: {x_test.shape}")
print(f"Shape of y_test: {y_test.shape}")
```

# scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)    #dataset is split into train & test in x we will have independent variables, y target features
x_test_scaled = scaler.transform(x_test)
```

# training the model


```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
# intialize the DecisionTree Classifier
DT = DecisionTreeClassifier()

# Train the model
DT.fit(x_train,y_train)
```

```python
# evaluate the model on test dataset
y_pred = DT.predict(x_test)
acs = accuracy_score(y_test, y_pred)
print('Accuracy Score', acs)
```

```python
# evaluate the model on test dataset
y_pred = DT.predict(x_test)
print(f"Accuracy score of Decision Tree Classifier is: {accuracy_score(y_test,y_pred)}")
```

```python
# evaluate the model on train dataset
y_pred_train = DT.predict(x_train)
print(f"Accuracy score of Decision Tree Classifier is: {accuracy_score(y_train,y_pred_train)}")
```

# Prediction

```python
def crop_rec(N, P, K, temp, hum, ph, rain):
    features = np.array([[N, P, K, temp, hum, ph, rain]])
    transformed_features = scaler.transform(features)
    prediction = DT.predict(transformed_features).reshape(1, -1)
    crop_dict = {
        1 : 'rice' ,
        2 : 'maize',
        3 : 'chickpea',
        4 : 'kidneybeans',
        5 : 'pigeonpeas',
        6 : 'mothbeans',
        7 : 'mungbean',
        8 : 'blackgram',
        9 : 'lentil',
        10: 'pomegranate',
        11 : 'banana',
        12 : 'mango',
        13 : 'grapes',
        14 : 'watermelon',
        15 : 'muskmelon',
        16 : 'apple',
        17 : 'orange',
        18 : 'papaya',
        19 : 'coconut',
        20 : 'cotton',
        21 : 'jute',
        22 : 'coffee'
    }
    crop = [crop_dict[i] for i in prediction[0]]
    return f"{crop} is a best crop to grow in the farm"
```

# Predicting with the values

```python
N = 90   
P = 45
K = 43
temp = 20
hum = 82
ph = 6.1
rain = 202

crop_rec(N, P, K, temp, hum, ph, rain)
```
